## Question 10
### Please import models and layers from the keras library.

```python
from keras import models, layers
```

## Question 11
### Define a sequential model and call it myNetwork.

```python
myNetwork = models.Sequential()
```

## Question 12
### Reshape the images from 28x28 to one column with 784 neurons (flattening)

```python
myNetwork.add(layers.Flatten(input_shape=(28, 28)))
```

## Question 13
### Also, please normalize the image by dividing the image by 255

```python
train_images_normalized = train_images / 255.0
test_images_normalized = test_images / 255.0
```

## Question 14
### Add one hidden layer that has 512 neurons, using the 'relu' activation function.

```python
myNetwork.add(layers.Dense(512, activation='relu'))
```

## Question 15
### Add another hidden layer that has 128 neurons, using the 'relu' activation function.

```python
myNetwork.add(layers.Dense(128, activation='relu'))
```

## Question 16
### Add the last layer as a 10-neuron dense layer that uses the 'softmax' as the activation function. Why do we use softmax for the last layer? How does it work under the hood? Make sure to explain this in your video.

```python
myNetwork.add(layers.Dense(10, activation='softmax'))
```

## Question 17 - Option A
### Use the following two settings for the compiler and run them separately and see what the differences are.
### Option A: Optimizer: adam, loss: 'sparse_categorical_crossentropy', metrics: ['accuracy']

```python
myNetwork.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

## Question 18 - Training with Option A
### Now after the compilation, please try to find the pattern using the fit command. Set the number of epochs to 10.

```python
history_optionA = myNetwork.fit(
    train_images_normalized,
    train_labels,
    epochs=10,
    validation_split=0.2
)
```

## Question 21 - Evaluation for Option A
### Use the evaluate() function from the Keras library to calculate the achieved accuracy and loss over the test images and labels. Do we have overfitting? Explain in your video.

```python
test_loss_A, test_accuracy_A = myNetwork.evaluate(test_images_normalized, test_labels)
print(f"Option A - Test Loss: {test_loss_A}")
print(f"Option A - Test Accuracy: {test_accuracy_A}")
```

## Question 17 - Option B
### Option B: Optimizer: rmsprop, loss: 'categorical_crossentropy', metrics: ['accuracy']

```python
from keras.utils import to_categorical

train_labels_categorical = to_categorical(train_labels)
test_labels_categorical = to_categorical(test_labels)

myNetwork_B = models.Sequential()
myNetwork_B.add(layers.Flatten(input_shape=(28, 28)))
myNetwork_B.add(layers.Dense(512, activation='relu'))
myNetwork_B.add(layers.Dense(128, activation='relu'))
myNetwork_B.add(layers.Dense(10, activation='softmax'))

myNetwork_B.compile(
    optimizer='rmsprop',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

## Question 18 - Training with Option B
### Now after the compilation, please try to find the pattern using the fit command. Set the number of epochs to 10.

```python
history_optionB = myNetwork_B.fit(
    train_images_normalized,
    train_labels_categorical,
    epochs=10,
    validation_split=0.2
)
```

## Question 21 - Evaluation for Option B
### Use the evaluate() function from the Keras library to calculate the achieved accuracy and loss over the test images and labels. Do we have overfitting? Explain in your video.

```python
test_loss_B, test_accuracy_B = myNetwork_B.evaluate(test_images_normalized, test_labels_categorical)
print(f"Option B - Test Loss: {test_loss_B}")
print(f"Option B - Test Accuracy: {test_accuracy_B}")
```

## Questions 19-20: Comparison

```python
print("\n=== COMPARISON ===")
print(f"Option A (adam + sparse_categorical_crossentropy):")
print(f"  Test Accuracy: {test_accuracy_A:.4f}")
print(f"  Test Loss: {test_loss_A:.4f}")
print(f"\nOption B (rmsprop + categorical_crossentropy):")
print(f"  Test Accuracy: {test_accuracy_B:.4f}")
print(f"  Test Loss: {test_loss_B:.4f}")
```

## Visualization of Training History

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history_optionA.history['accuracy'], label='Option A Train')
plt.plot(history_optionA.history['val_accuracy'], label='Option A Val')
plt.plot(history_optionB.history['accuracy'], label='Option B Train')
plt.plot(history_optionB.history['val_accuracy'], label='Option B Val')
plt.title('Model Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history_optionA.history['loss'], label='Option A Train')
plt.plot(history_optionA.history['val_loss'], label='Option A Val')
plt.plot(history_optionB.history['loss'], label='Option B Train')
plt.plot(history_optionB.history['val_loss'], label='Option B Val')
plt.title('Model Loss Comparison')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
